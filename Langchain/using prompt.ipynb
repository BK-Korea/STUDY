```python
from langchain_ollama import ChatOllama
# 1. LLM 객체 생성
llm = ChatOllama(model="llama3.2")
llm.invoke("waht is the capital of korea?")
```

```python
from langchain_core.prompts import PromptTemplate

# 2. 프롬프트 템플릿 정의
Prompt_Template = PromptTemplate(
    template="what is the capital of {country}?",
    input_variables=["country"],
)

# 3. 템플릿에 값 채워서 프롬프트 생성
prompt = Prompt_Template.invoke({"country": "france"})
print(prompt)

# 4. 첫 번째 방식: 만들어 놓은 프롬프트를 LLM에게 전달
llm.invoke(prompt)

# 5. 두 번째 방식: 템플릿을 한 번에 치환하고 LLM에게 전달
#llm.invoke(Prompt_Template.invoke({"country": "france"}))
```

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage

#답변 에시임, 논문 few shot learning, openai 논문에 따르면 이런걸 주는게 훨씬 정확도 좋음
message_list = [
    SystemMessage(content="You are a helpful assistant"),
    HumanMessage(content="what is the capital of korea?"),
    AIMessage(content="The capital of korea is seoul."),
    HumanMessage(content="what is the capital of france?"),
    AIMessage(content="The capital of france is paris."),
]

llm.invoke(message_list)

#이건 그러나 좀 langchain스럽지 않음
```

```python
#좀 더 langchain스러운 방법, 채팅 메시지 처럼 쓴다.

from langchain_core.prompts import ChatPromptTemplate

#message list가 아니라 튜플로 넣어야함 그리고 이게 좀 더 랭체인스러운 방법이라 함
#LCEL을 이용할 수 있는 구조
chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant"),
    ("human", "what is the capital of {country}?"),
])

chat_prompt = chat_prompt_template.invoke({"country": "france"})

print(chat_prompt)

```

```python
#llm은 GPT 모델을 연결한 langchain 객체
#invoke는 질문의 대한 답을 얻는 것 

llm.invoke(chat_prompt)
```
